```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_train = np.array([2,4,6,8,9])
y_train = np.array([200, 55, 90, 44, 120])
print(f"x_train is : {x_train.shape[0]}")
print(f"y_train is : {y_train.shape[0]}")
```

```python
#m is the number of tarining samples
m = x_train.shape[0]
print(m) 
```

```python
i = 0
x_i = x_train[i]
y_i = y_train[i]
print(f"(x^{i} , y^{i}) : ({x_i} , {y_i})")
```

```python
plt.scatter(x_train , y_train , marker='x' , c='r')
plt.title("House Prediction")
plt.ylabel("Age")
plt.xlabel("Number of People")
```

```python
w = 100
b = 100
print(f"weight is : {w} and bias is : {b}")
```

```python
def compute_model_output(x,w,b):
    m = x.shape[0]
    fw_b = np.zeros(m)
    for i in range(m):
        fw_b[i] = w * x[i] + b
    return fw_b
    
    
```

```python
temp_f_wb = compute_model_output(x_train , w , b)
#Plot model prediction
plt.plot(x_train , temp_f_wb , label = 'Our Prediction', c='r')
plt.scatter(x_train , y_train, marker='x' , c = 'b' , label = "Actual Values")
plt.title("Model Training")
plt.xlabel("Number of People")
plt.ylabel("Age")
plt.legend()
plt.show()
```
