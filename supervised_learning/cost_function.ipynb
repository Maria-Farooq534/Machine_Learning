```python
x_train = np.array([1.0 , 2.0])
y_train = np.array([200.0 , 300.0])
```

Step 2: Define Cost Function

```python
def compute_cost(x,y,w,b):
    m = x.shape[0]
    total_cost = 0

    for each_sample in range (m):
        prediction = w * x[each_sample] + b
        error = prediction - y[each_sample]
        sample_cost = error ** 2
        total_cost = sample_cost + total_cost


    return total_cost / (2*m)
```

Step 3: Try with some values of w and b

```python
w = 100
b = 100
cost = compute_cost(x_train , y_train , w , b)
print(f"Cost of w {w} and b {b} is : {cost}")
```

```python
cost2 = compute_cost(x_train , y_train , 100 , 100)
print(f"Cost is : {cost2}")
```
